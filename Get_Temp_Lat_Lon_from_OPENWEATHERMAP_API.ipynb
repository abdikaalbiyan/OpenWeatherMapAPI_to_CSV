```python
import requests
import csv


API_key = 'YOUR_API_KEY'
base_url = "http://api.openweathermap.org/data/2.5/weather?"

weather_data_dict = {}
with open('/mnt/c/Belajar/Data/kota_indonesia.csv', 'r') as reader_file:
    reader = csv.reader(reader_file)
    for row in reader:
        city_name = row[0]

        final_url = base_url + "appid=" + API_key + "&q=" + city_name
        
        weather_data = requests.get(final_url).json()
        
        if weather_data['cod'] != '404':
            weather_data_dict[weather_data['name']] = [(weather_data['main']['temp'] - 273.15), weather_data['coord']['lat'], weather_data['coord']['lon']]
        
        
 
```

```python
for key, value in weather_data_dict.items():
    data = [
        key,
        value[0],
        value[1],
        value[2]
    ]
    print(data)
```

Write CSV

```python
with open('/mnt/c/Belajar/kota_indonesia_plus_temperatur.csv', 'w') as output_file:
    csv_writer = csv.writer(output_file)
    csv_writer.writerow(['Kota', 'Temperatur', 'Latitude', 'Longitude'])
    for key, value in weather_data_dict.items():
        data = [
            key,
            value[0],
            value[1],
            value[2]
        ]
        csv_writer.writerow(data)
```
